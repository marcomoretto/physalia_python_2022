### Import libraries

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

### Read the dataset and put it into a DataFrame

```python
df = pd.read_csv('/content/limma-voom_luminalpregnant-luminallactate', sep='\t', index_col=0)
df.head()
```

### Set thresholds for logFC and p-value and adjust the DF adding log10 transformation and adding an extra column for DE and non-DE genes

```python
log_fc_threshold = 2.0
p_value_threshold = 0.05

df = df[['SYMBOL', 'logFC', 'adj.P.Val']]
df.loc[:,'group'] = 'Not DE genes'
df.loc[(df['logFC'] >= log_fc_threshold)&(df['adj.P.Val'] <= p_value_threshold), 'group'] = 'DE genes'
df.loc[(df['logFC'] <= -log_fc_threshold)&(df['adj.P.Val'] <= p_value_threshold), 'group'] = 'DE genes'
df['adj.P.Val'] = -np.log10(df['adj.P.Val'])
df.head()
```

### Volcano plot

```python
fig = px.scatter(
    df,
    x='logFC',
    y='adj.P.Val',
    color='group',
    hover_name="SYMBOL", 
    hover_data=["logFC", "adj.P.Val"],
    width=600, 
    height=800,
    color_discrete_map={
        'DE genes': 'red', 
        'Not DE genes': 'gray'
    }
)
fig
```
