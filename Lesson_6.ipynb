```python
alt.Chart(p_df.dropna()).transform_density(
    'body_mass_g', 
    groupby=['sex', 'species'],
    as_=['body_mass_g', 'density']
).mark_area(opacity=0.5).encode(
    x='body_mass_g:Q',
    y='density:Q',
    color='sex:N',
    column='species:N'
).properties(
    width=400,
    height=300
)
```

```python
def ecdf(df, row):
  same_sex_species_df = df[(df['sex'] == row['sex'])&(df['species'] == row['species'])]
  try:
    return same_sex_species_df[same_sex_species_df['body_mass_g'] <= row['body_mass_g']].shape[0] / same_sex_species_df.shape[0]
  except Exception as e:
    return np.nan

ecdf_p_df = p_df.copy()

ecdf_p_df['ecdf'] = ecdf_p_df.sort_values(by=['sex', 'species', 'body_mass_g']).apply(
    lambda x : ecdf(ecdf_p_df, x),
    axis=1
)

ecdf_p_df.head(10)
```

```python
alt.Chart(ecdf_p_df.dropna()).mark_line(opacity=0.5).encode(
    x='body_mass_g',
    y='ecdf',
    column='sex',
    color='species'
).properties(
    width=600,
    height=400
)
```
